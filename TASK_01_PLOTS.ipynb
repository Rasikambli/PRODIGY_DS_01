```python
plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.title('Sales_in_thousands')
data['Sales_in_thousands'].hist(bins=15,edgecolor='Black',alpha=0.5,color='Red')
plt.subplot(1,3,2)
plt.title('__year_resale_value')
data['__year_resale_value'].hist(bins=15,edgecolor='Black',alpha=0.5,color='Pink')
plt.subplot(1,3,3)
plt.title('Horsepower')
data['Horsepower'].hist(bins=15,edgecolor='Black',alpha=0.5,color='Purple')

plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.title('Engine_size')
data['Engine_size'].hist(bins=15,edgecolor='Black',alpha=0.5,color='Green')
plt.subplot(2,3,2)
plt.title('Fuel_capacity')
data['Fuel_capacity'].hist(bins=15,edgecolor='Black',alpha=0.5,color='Orange')
plt.subplot(2,3,3)
plt.title('Fuel_efficiency')
data['Fuel_efficiency'].hist(bins=15,edgecolor='Black',alpha=0.5,color='brown')
```
